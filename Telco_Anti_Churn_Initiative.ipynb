## Transforming the data


```python
# drop rows of column 'Total Charges' that are not numerical
# convert 'Total Charges' column from strings to floats

import re

rows = []
for index, row in df.iterrows():
  if not bool(re.match('^[0-9|.]+$', row['Total Charges'])):
    rows.append(index)

df.drop(index=rows, inplace=True)

df['Total Charges'] = df['Total Charges'].apply(lambda x: float(x))
```

```python
df.columns
```

### One-Hot Encoding Categorical Data
Many of the columns in the data set were categorical and this required us to use one-hot encoding. This transformed our data into something our models could use in training.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

ohe = OneHotEncoder(sparse=False)

transformer = make_column_transformer(
    (ohe, ['Gender', 'Senior Citizen', 'Partner', 'Dependents', 'Phone Service', 'Multiple Lines', 'Internet Service', 
                      'Online Security', 'Online Backup', 'Device Protection', 'Tech Support', 'Streaming TV', 'Streaming Movies',
                      'Contract', 'Paperless Billing', 'Payment Method']),
                      remainder='passthrough'
)
transformer.set_output(transform='pandas')
transformed_df = transformer.fit_transform(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

hash = {}

for i in df.index:
  pm = df['Payment Method'][i]
  if pm in hash:
    hash[pm] += 1
  else:
    hash[pm] = 1

my_labels = hash.keys()
values = list(map(lambda x: hash[x], my_labels))
y = np.array(values)

#Making the Plot
colors = ['#a7bed3','#c6e2e9','#f1ffc4','#ffcaaf']
plt.title('Payment Methods for Entire Data Set')
plt.pie(y, labels=my_labels, colors = colors)
plt.show()
```

```python
# create new dictionary
new_hash = {}
for l in my_labels:
  new_hash[l] = [0, 0]

# record churn values
for i in df.index:
  pm = df['Payment Method'][i]
  cv = df['Churn Value'][i]
  churned = True if cv == 1 else False
  if churned:
    new_hash[pm][0] += 1
  else:
    new_hash[pm][1] += 1

# compute churn rates
def compute_cr (pm):
  counts = new_hash[pm]
  churn_count = counts[0]
  stay_count = counts[1]
  return churn_count / (churn_count + stay_count)

# y = list(map(compute_cr, my_labels))
y = list(map(lambda label: compute_cr(label), my_labels))

plt.title('Payment Methods for Churned Customers')
plt.xlabel("Payment Methods")
plt.ylabel("Churn Rate")

plt.bar(my_labels, y)
plt.tick_params(rotation=45)
plt.show()
```

```python
def graph_cr(df, feature):
  hash = {}
  for i in df.index:
    val = df[feature][i]
    if not val in hash:
      hash[val] = [0, 0]
    counts = hash[val]
    cv = df['Churn Value'][i]
    if cv == 1:
      counts[0] += 1
    else:
      counts[1] += 1
    hash[val] = counts
  my_labels = hash.keys()
  def compute_cr(label):
    counts = hash[label]
    churned = counts[0]
    not_churned = counts[1]
    return churned / (churned + not_churned)
  y = list(map(lambda label: compute_cr(label), my_labels))
  plt.title(f'{feature} for churned customers')
  plt.xlabel(feature)
  plt.ylabel('Churn rate')
  plt.bar(my_labels, y)
  plt.tick_params(rotation=45)
  plt.show()


```

```python
from scipy.stats import pearsonr

def graph_corr(df, feature):
  x = []
  y = []
  for i in df.index:
    x.append(df[feature][i])
    y.append(df['Churn Score'][i])
  plt.scatter(x, y)
  plt.xlabel(feature)
  plt.ylabel('Churn score')
  plt.title(f'{feature} versus churn score')
  plt.show()
  print(f'Correlation coefficient: {pearsonr(x, y)[0]}')
```

```python
transformed_df = transformed_df.drop(columns=['remainder__Churn Reason', 'remainder__Churn Label','remainder__Churn Score','remainder__Lat Long','remainder__Country', 'remainder__State', 'remainder__City', 'remainder__Zip Code', 'remainder__Count', 'remainder__Latitude', 'remainder__Longitude'])
```

```python
y = transformed_df['remainder__Churn Value']
transformed_df = transformed_df.drop(columns=['remainder__Churn Value'])
```

```python
X = transformed_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFECV

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

rfe_method = RFECV(
    RandomForestClassifier()
)

rfe_method.fit(X_train, y_train)

X_train.columns[(rfe_method.get_support())]
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Fit a random forest classifier on the training data
rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train)

# Select the top 3 features based on the feature importances of the random forest
sfm = SelectFromModel(rfc, threshold=-np.inf, max_features=10)
sfm.fit(X_train, y_train)

# Get the top 3 features
top_features = X_train.columns[sfm.get_support(indices=True)].tolist()

print(f"The top 10 most important features are:")

for n in top_features:
  print(n)
```

```python
sel_columns = X_train.columns[(rfe_method.get_support())]
```

```python
X_train_selected = X_train[sel_columns]
X_test_selected = X_test[sel_columns]
```

# Training
> We will be training three different models:
- Random Forest
- Naive Bayes
- Logistic Regression
- Multilayer Perceptron

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

def print_metrics(y_test, y_test_pred):
  print(f'Accuracy: {accuracy_score(y_test, y_test_pred)}')
  print(f'Precision: {precision_score(y_test, y_test_pred)}')
  print(f'Recall: {recall_score(y_test, y_test_pred)}')
  print(f'F1: {f1_score(y_test, y_test_pred)}')
```

```python
rf_classifier = RandomForestClassifier()

rf_classifier.fit(X_train_selected, y_train)

y_test_pred_rf = rf_classifier.predict(X_test_selected)

print_metrics(y_test, y_test_pred_rf)
```

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()

nb_classifier.fit(X_train_selected, y_train)

y_test_pred_nb = nb_classifier.predict(X_test_selected)

print_metrics(y_test, y_test_pred_nb)
```

```python
train_features = X_train_selected
train_labels = y_train


# Now let's define our models
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier


lr_classifier = LogisticRegression(solver='lbfgs',max_iter=10000)
mlp_classifier = MLPClassifier(solver='lbfgs', alpha=1e-5,
                               hidden_layer_sizes=(8, 2), random_state=11,max_iter=10000)


# train our models
lr_classifier.fit(train_features.to_numpy(),train_labels.to_numpy())
mlp_classifier.fit(train_features.to_numpy(),train_labels.to_numpy())
```

## Testing Accuracy of Models

```python
from sklearn.metrics import accuracy_score

y_actual = y_test
test_inputs = X_test_selected

# predict using logistic regression model
y_predicted_lr = lr_classifier.predict(test_inputs.to_numpy())

print_metrics(y_actual, y_predicted_lr)
```

```python
# predict using logistic regression model
y_predicted_mlp = mlp_classifier.predict(test_inputs.to_numpy())

print_metrics(y_actual, y_predicted_mlp)
```
